**For Scatter Graph**

```python
highest_rated=data.sort_values(by="Star Rating",ascending=False)##default sort value is ascending so we have made it false for descending
X=highest_rated["Sale Price"]
Y=highest_rated["Mrp"]
fig=go.Figure(go.Scatter(x=X,y=Y,mode="markers"))
fig.update_layout(title="Sale Price vs Mrp Graph",xaxis_title="X-Axis",yaxis_title="Y-Axis",)
fig.show()
```

**For bar Graph and For getting top 10 datas data visualization**

```python
highest_rated=data.sort_values(by="Star Rating",ascending=False)##default sort value is ascending so we have made it false for descending
highest_rated1=highest_rated.head(10)#For getting only 10 datas
X=highest_rated1["Product Name"]
Y=highest_rated1["Mrp"]
fig=px.bar(x=X,y=Y,title="Sales Price vs Mrp Graph of Iphones")
fig.show()
```

**For getting all the datas data visualization**

```python
highest_rated=data.sort_values(by="Star Rating",ascending=False)##default sort value is ascending so we have made it false for descending
X=highest_rated["Product Name"]
Y=highest_rated["Mrp"]
fig=px.bar(x=X,y=Y,title="Sales Price vs Mrp Graph of Iphones")
fig.show()
```

```python
import plotly.express as px
highest_rated=data.sort_values(by="Star Rating",ascending=False)##default sort value is ascending so we have made it false for descending
for_10_datasets=highest_rated.head(10)
X=for_10_datasets["Product Name"]
Y=for_10_datasets["Number Of Reviews"]
figure=px.bar(x=X,y=Y,title="Name and Reviews of Iphones")
figure.show()

```

```python
highest_rated=data.sort_values(by="Product Name",ascending=False)
for_10_datasets=highest_rated.head(10)
X=for_10_datasets["Product Name"]
Y=for_10_datasets["Sale Price"]
fig=px.bar(x=X,y=Y,title="Sales vs Names of Iphones")
fig.show()
```

```python
X=highest_rated["Product Name"]
Y=highest_rated["Sale Price"]
Z=highest_rated["Discount Percentage"]
figure=go.Figure(go.Scatter(x=X,y=Y,mode="markers"))
figure.update_layout(xaxis_title="Product Name",yaxis_title="Sale Price",title="Relationship between sales price and name of Iphones")
figure.show()
```

**For 10 Datasets**

```python
highest_rated=data.sort_values(by="Product Name",ascending=False)
for_10_datasets=highest_rated.head(10)
X=for_10_datasets["Number Of Ratings"]
Y=for_10_datasets["Sale Price"]
figure=go.Figure(go.Scatter(x=X,y=Y,mode="markers"))
figure.update_layout(xaxis_title="Number Of Ratings",yaxis_title="Sale Price",title="Relationship between sales price and Number Of Ratings of Iphones")
figure.show()
```

```python
figure=px.scatter(data_frame=data,x="Number Of Ratings",y="Sale Price",size="Discount Percentage",trendline="ols",title="Relationship between sales price and Number Of Ratings of Iphones")
figure.show()
```
